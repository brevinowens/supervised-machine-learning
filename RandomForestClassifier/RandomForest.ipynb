```python
#dependencies

import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
```

```python
#importing csv file

lend = pd.read_csv('../Resources/lending_data.csv')
lend.head(10)
```

```python
#declaring X & y variables

X = lend.drop("loan_status", axis=1)
y = lend['loan_status']
```

```python
#preparing the data

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#fitting the model

lend_rf = RandomForestClassifier(random_state=42).fit(X_train_scaled, y_train)
y_pred = lend_rf.predict(X_test_scaled)
```

```python
#scoring the model and printing a classification report

print(classification_report(y_test, y_pred))
print(f'Training Score: {lend_rf.score(X_train_scaled, y_train)}')
print(f'Testing Score: {lend_rf.score(X_test_scaled, y_test)}')
```

The scores for both models are .99 and above, but I still believe that the Random Forest Classifier model better explains the data. This is because the variables do not have a linear relationship, which makes me question why we are recieving such a high testing score with the Logistic Regression model.
